```python
import torch
import torch.nn as nn
import torchvision.models as models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def build_model(model_name):
    if model_name == "resnet50":
        model = models.resnet50(weights=None)
        model.fc = nn.Linear(model.fc.in_features, 2)

    elif model_name == "efficientnet_b3":
        model = models.efficientnet_b3(weights=None)
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)

    elif model_name == "vit_b_16":
        model = models.vit_b_16(weights=None)
        model.heads.head = nn.Linear(model.heads.head.in_features, 2)

    return model.to(device)



resnet_model = build_model("resnet50")
efficient_model = build_model("efficientnet_b3")
vit_model = build_model("vit_b_16")

resnet_model.load_state_dict(torch.load("resnet.pt", map_location=device))
efficient_model.load_state_dict(torch.load("efficient.pt", map_location=device))
vit_model.load_state_dict(torch.load("vit.pt", map_location=device))


for model in [resnet_model, efficient_model, vit_model]:
    for param in model.parameters():
        param.requires_grad = False


resnet_model.eval()
efficient_model.eval()
vit_model.eval()

print(" Models loaded & frozen")
```

```python
features_store = {}

def get_hook(name):
    def hook(module, input, output):
        if name == "vit":

            features_store[name] = output[:, 0, :]
        else:

            features_store[name] = output.mean(dim=[2, 3])
    return hook


resnet_model.layer4.register_forward_hook(get_hook("resnet"))
efficient_model.features.register_forward_hook(get_hook("efficient"))
vit_model.encoder.register_forward_hook(get_hook("vit"))

print("Hooks registered successfully")
```

```python
import torchvision.transforms as transforms
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.3),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

print(" Transforms ready")
```

```python
from torch.utils.data import Dataset
from PIL import Image
import os

class BreastCancerDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.image_paths = []
        self.labels = []
        self.transform = transform

        for patient in os.listdir(root_dir):
            patient_path = os.path.join(root_dir, patient)

            if not os.path.isdir(patient_path):
                continue

            for class_folder in ["0", "1"]:
                class_path = os.path.join(patient_path, class_folder)

                if not os.path.exists(class_path):
                    continue

                for img in os.listdir(class_path):
                    if img.endswith(".png"):
                        self.image_paths.append(os.path.join(class_path, img))
                        self.labels.append(int(class_folder))

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        label = self.labels[idx]

        if self.transform:
            img = self.transform(img)

        return img, label
```

```python
from torch.utils.data import random_split, DataLoader, WeightedRandomSampler
import numpy as np


train_size = int(0.7 * len(dataset))
val_size   = int(0.15 * len(dataset))
test_size  = len(dataset) - train_size - val_size

train_dataset, val_dataset, test_dataset = random_split(
    dataset, [train_size, val_size, test_size]
)

print("Train:", len(train_dataset))
print("Val:", len(val_dataset))
print("Test:", len(test_dataset))
```

```python

train_labels = [dataset.labels[i] for i in train_dataset.indices]

class_counts = np.bincount(train_labels)
class_weights = 1. / class_counts


sample_weights = [class_weights[label] for label in train_labels]

sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)

print("Class counts:", class_counts)
print("Class weights:", class_weights)
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, sampler=sampler)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)

print("✅ DataLoaders ready")
```

```python
import numpy as np
from tqdm import tqdm

def extract_features(loader):
    all_features = []
    all_labels = []

    with torch.no_grad():
        for x, y in tqdm(loader):
            x = x.to(device)


            resnet_model(x)
            efficient_model(x)
            vit_model(x)

            f1 = features_store["resnet"]
            f2 = features_store["efficient"]
            f3 = features_store["vit"]
            fused = torch.cat([f1, f2, f3], dim=1)

            all_features.append(fused.cpu())
            all_labels.extend(y.numpy())

    return torch.cat(all_features), np.array(all_labels)
```

```python
features, labels = extract_features(test_loader)

print("Feature shape:", features.shape)
```

```python
import torch.nn as nn

class FusionClassifier(nn.Module):
    def __init__(self):
        super().__init__()

        self.net = nn.Sequential(
            nn.Linear(4352, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(512, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.2),

            nn.Linear(128, 2)
        )

    def forward(self, x):
        return self.net(x)


fusion_model = FusionClassifier().to(device)

print(" Fusion model ready")
```

```python
from torch.utils.data import TensorDataset, DataLoader

features = features.float()
labels = torch.tensor(labels)

fusion_dataset = TensorDataset(features, labels)
fusion_loader = DataLoader(fusion_dataset, batch_size=64, shuffle=True)

print("Fusion data ready")
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(fusion_model.parameters(), lr=1e-3)

for epoch in range(5):
    fusion_model.train()
    total_loss = 0

    for x, y in fusion_loader:
        x, y = x.to(device), y.to(device)

        optimizer.zero_grad()
        out = fusion_model(x)
        loss = criterion(out, y)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix

fusion_model.eval()

all_preds = []
all_probs = []

with torch.no_grad():
    for x, y in fusion_loader:
        x = x.to(device)

        out = fusion_model(x)
        probs = torch.softmax(out, dim=1)

        preds = torch.argmax(out, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_probs.extend(probs[:,1].cpu().numpy())


acc = accuracy_score(labels, all_preds)
f1  = f1_score(labels, all_preds)
auc = roc_auc_score(labels, all_probs)
cm  = confusion_matrix(labels, all_preds)

print("Accuracy:", acc)
print("F1 Score:", f1)
print("AUC-ROC:", auc)
print("Confusion Matrix:\n", cm)
```

```python
import torchvision.transforms as transforms

test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
from torch.utils.data import Dataset
from PIL import Image
import os

class BreastCancerDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.image_paths = []
        self.labels = []
        self.transform = transform

        for patient in os.listdir(root_dir):
            patient_path = os.path.join(root_dir, patient)

            if not os.path.isdir(patient_path):
                continue

            # 🔥 IMPORTANT: go inside class folders
            for label in ["0", "1"]:
                class_path = os.path.join(patient_path, label)

                if not os.path.exists(class_path):
                    continue

                for img_name in os.listdir(class_path):
                    if img_name.endswith(".png"):
                        self.image_paths.append(os.path.join(class_path, img_name))
                        self.labels.append(int(label))

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert("RGB")
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label
```

```python
img, _ = dataset[0]

input_tensor = img.unsqueeze(0).to(device)
```

```python
import torch
import torch.nn as nn
import torchvision.models as models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def build_model(model_name):
    if model_name == "resnet50":
        model = models.resnet50(weights="DEFAULT")
        model.fc = nn.Linear(model.fc.in_features, 2)

    elif model_name == "efficientnet_b3":
        model = models.efficientnet_b3(weights="DEFAULT")
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)

    elif model_name == "vit_b_16":
        model = models.vit_b_16(weights="DEFAULT")
        model.heads.head = nn.Linear(model.heads.head.in_features, 2)

    return model.to(device)
```

```python
vit_model = build_model("vit_b_16")
vit_model.load_state_dict(torch.load("vit.pt", map_location=device))
vit_model.to(device)
vit_model.eval()
```

```python
attentions = []

x = input_tensor
x = vit_model._process_input(x)
n = x.shape[0]

batch_class_token = vit_model.class_token.expand(n, -1, -1)
x = torch.cat([batch_class_token, x], dim=1)

x = x + vit_model.encoder.pos_embedding
x = vit_model.encoder.dropout(x)

for layer in vit_model.encoder.layers:
    x_norm = layer.ln_1(x)
    attn_output, attn_weights = layer.self_attention(
        x_norm, x_norm, x_norm, need_weights=True
    )

    attentions.append(attn_weights.detach().cpu())

    x = x + layer.dropout(attn_output)
    x = x + layer.mlp(layer.ln_2(x))

print("Captured layers:", len(attentions))
```

```python
import torch

def compute_rollout(attentions):
    result = torch.eye(attentions[0].size(-1))

    for attn in attentions:
        attn = attn.mean(dim=1)  # average heads

        # add residual connection
        attn = attn + torch.eye(attn.size(-1))

        # normalize
        attn = attn / attn.sum(dim=-1, keepdim=True)

        result = torch.matmul(result, attn)

    return result
```

```python
rollout = compute_rollout(attentions)

mask = rollout[0, 1:]
mask = mask.reshape(14, 14).numpy()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

mask = cv2.resize(mask, (224, 224))
mask = (mask - mask.min()) / (mask.max() - mask.min())

img_np = input_tensor[0].permute(1,2,0).cpu().numpy()
img_np = (img_np - img_np.min()) / (img_np.max() - img_np.min())

heatmap = cv2.applyColorMap(np.uint8(255 * mask), cv2.COLORMAP_JET)
heatmap = heatmap / 255

overlay = heatmap * 0.5 + img_np * 0.5

plt.imshow(overlay)
plt.title("ViT Attention Rollout")
plt.axis('off')
```

```python
import numpy as np

def region_analysis(mask, model_name="Model"):
    flat = mask.flatten()

    # Top 10 important regions
    top_idx = np.argsort(flat)[-10:]

    # Mean importance
    importance = np.mean(flat)

    # High attention %
    threshold = 0.6
    high_attention = flat > threshold
    percentage = np.sum(high_attention) / len(flat) * 100

    print(f"\n🔹 {model_name}")
    print("Top patch indices:", top_idx)
    print("Mean importance:", round(importance, 4))
    print("High-attention %:", round(percentage, 2), "%")

    return {
        "top_idx": top_idx,
        "importance": importance,
        "percentage": percentage
    }
```
